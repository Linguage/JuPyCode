```python
import autograd.numpy as np  # Thinly-wrapped numpy
from autograd import grad    # The only autograd function you may ever need
```

```python
def tanh(x):                 # Define a function
     y = np.exp(-2.0 * x)
     return (1.0 - y) / (1.0 + y)
```

```python
grad_tanh = grad(tanh)       # Obtain its gradient function
grad_tanh(1.0)               # Evaluate the gradient at x = 1.0
```

```python
(tanh(1.0001) - tanh(0.9999)) / 0.0002  # Compare to finite differences
```

```python
from __future__ import absolute_import
import autograd.numpy as np
import matplotlib.pyplot as plt
from autograd import elementwise_grad as egrad

'''
Mathematically we can only take gradients of scalar-valued functions, but
autograd's elementwise_grad function also handles numpy's familiar vectorization
of scalar functions, which is used in this example.

To be precise, elementwise_grad(fun)(x) always returns the value of a
vector-Jacobian product, where the Jacobian of fun is evaluated at x and the
vector is an all-ones vector with the same size as the output of fun. When
vectorizing a scalar-valued function over many arguments, the Jacobian of the
overall vector-to-vector mapping is diagonal, and so this vector-Jacobian
product simply returns the diagonal elements of the Jacobian, which is the
(elementwise) gradient of the function at each input value over which the
function is vectorized.
'''

def tanh(x):
    return (1.0 - np.exp(-x))  / (1.0 + np.exp(-x))

x = np.linspace(-7, 7, 200)
plt.plot(x, tanh(x),
         x, egrad(tanh)(x),                                     # first derivative
         x, egrad(egrad(tanh))(x),                              # second derivative
         x, egrad(egrad(egrad(tanh)))(x),                       # third derivative
         x, egrad(egrad(egrad(egrad(tanh))))(x),                # fourth derivative
         x, egrad(egrad(egrad(egrad(egrad(tanh)))))(x),         # fifth derivative
         x, egrad(egrad(egrad(egrad(egrad(egrad(tanh))))))(x))  # sixth derivative

plt.axis('off')
plt.savefig("tanh.png")
plt.show()
```

```python
from __future__ import absolute_import
import autograd.numpy as np
import matplotlib.pyplot as plt
from autograd import elementwise_grad as egrad

def exp2(x):
    return np.sin(x)+x**(-1)

x = np.linspace(1, 7, 200)

plt.plot(x, exp2(x),
         x, egrad(exp2)(x),
         x, egrad(egrad(exp2))(x),
         x, egrad(egrad(egrad(exp2)))(x))  # sixth derivative

# plt.axis('off')
plt.savefig("exp2.png")
plt.show()
```

```python
# 导入所需的库
from autograd import numpy as np
from autograd import elementwise_grad as egrad
from autograd import jacobian
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 定义谐振子方程的函数
def harmonic_oscillator(t, y, k):
    # y 是一个包含两个元素的数组，分别是位移 y 和速度 dy/dt
    position, velocity = y
    
    # 定义被求导的函数
    def func(y):
        # 返回 [velocity, -k * position] 即一阶微分方程的导数
        return np.array([velocity, -k * position])
    
    # 使用 autograd 计算函数的一阶导数
    grad = jacobian(func)(y)
    
    # 返回 dy/dt 和 ddy/dt（自动微分计算的）
    return func(y)

# 定义常数 k
k = 1.0  # 可以根据需要调整

# 初始条件
y0 = [1.0, 0.0]  # 初始位移为 1，初始速度为 0

# 时间范围
t_span = (0, 10)  # 从 0 到 10 的时间范围

# 使用 solve_ivp 求解二阶微分方程
solution = solve_ivp(lambda t, y: harmonic_oscillator(t, y, k), t_span, y0, method='RK45', t_eval=np.linspace(0, 10, 100))

# 绘制结果
plt.figure()
plt.plot(solution.t, solution.y[0], label='Position (y)')
plt.plot(solution.t, solution.y[1], label='Velocity (dy/dt)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Harmonic Oscillator')
plt.legend()
plt.show()
```
