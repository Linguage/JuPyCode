```julia
# fibonacc数列计算

function fibonacci(n)
    if n <= 1
        return n

    else
        return fibonacci(n-1) + fibonacci(n-2)
    end
end

# 打印前10个数落值

for i in 1:10
    print(fibonacci(i), " ")
end

# 绘制前100个Fibonacci数列的曲线

# using Plots

# x = [i for i in 1:100]
# y = [fibonacci(i) for i in 1:100]
#
# plot(x, y, label="Fibonacci", title="Fibonacci Series", xlabel="n", ylabel="Fibonacci(n)")

# 绘制斐波那契数列的黄金分割线

# plot!([0, 100], [0, 0], label="Golden Ratio", color="red", linestyle=:dash)

```

```julia
# 自定义一个函数，输入可以为数值、向量或矩阵

function my_func(x)
    if isa(x, Number)
        return x^2

    elseif isa(x, Array)
        return x.^2
    else
        error("Input must be a number or an array")

    end
end


# 测试
my_func(2) # 4
my_func([1, 2, 3]) # [1, 4, 9]
# my_func("hello") # ERROR: Input must be a number or an array



```

```julia
# 使用julia定义函数，求解向量的三角函数值

# using LinearAlgebra

function trig_func(x)
    return [sin(x), cos(x), tan(x)]
end

x = [0, pi/4, pi/2, 3pi/4, pi]

for i in 1:length(x)
    print("The value of sin(x), cos(x), tan(x) at x = ", x[i], " is: ")
    println(trig_func(x[i]))
end
```

```julia
# 使用julia定义函数，求解向量的sin值

using LinearAlgebra
using Plots

function sin_vec(x)
    return sin.(x)
end

x = 0:0.1:2*pi
y = sin_vec(x)


plot(x, y, label="sin(x)")

xlabel!("x")
ylabel!("y")
title!("sin(x) Curve")
# println(y)
# [0.8414709848078965, 0.9092974268256817, 0.1411200080598672]
```

```julia
# 使用julia定义函数，求解向量的sin值

using LinearAlgebra
using ForwardDiff
using Plots

# 定义向量函数
function sin_vec(x)
    return sin.(x)
end

# 定义 x 的范围
x = 0:0.1:5*π

# 计算 y = sin(x)
y = sin_vec(x)

# 使用 jacobian 计算 sin_vec 在 x 上的导数
df = diag(ForwardDiff.jacobian(sin_vec, x));


# # 绘制函数和导数曲线
plot(x, y, label="sin(x)", xlabel="x", ylabel="y", title="Function and Derivative of sin(x)")
plot!(x, df, label="cos(x)")
# # legend()
# show()
```
